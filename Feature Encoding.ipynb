```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "color":["red","green","blue","green","red"],
    "size":["small","medium","large","medium","small"],
    "city":["hyderabad","delhi","mumbai","delhi","chennnai"],
    "target":[10,20,30,40,50]})
```

```python
df
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df["color_Label"] = LabelEncoder().fit_transform(df["color"])
df[["color","color_Label"]]
```

```python
ohe = pd.get_dummies(df, columns=["color"])
print(ohe)
```

# One- Hot Encoding

```python
one_hot=pd.get_dummies(df["color"],prefix="color")
one_hot
```

```python
oh=pd.get_dummies(df,columns=["color"])
oh
```

# Target Encoding

```python
te= df.groupby("city")["target"].mean()

df["city_target_Enc"]=df["city"].map(te)
df[["city", "city_target_Enc"]]
```

# Frequency Encoding

```python
freq = df["city"].value_counts()
df["city_freq_Enc"] = df["city"].map(freq)
df[["city","city_freq_Enc"]]
```
